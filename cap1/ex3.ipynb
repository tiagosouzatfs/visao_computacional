```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carregar as imagens
img1 = plt.imread('tropical1.png')
img2 = plt.imread('tropical2.png')
```

```python

# Verificar se as imagens têm o mesmo tamanho
if img1.shape != img2.shape:
    raise ValueError("As imagens devem ter o mesmo tamanho para adição.")
```

```python
# Adicionar as imagens
added_img = img1 + img2
```

```python
# Exibir as imagens lado a lado
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

# Exibir a primeira imagem
axs[0].imshow(img1)
axs[0].set_title('Imagem 1')
axs[0].axis('off')

# Exibir a segunda imagem
axs[1].imshow(img2)
axs[1].set_title('Imagem 2')
axs[1].axis('off')

# Exibir a imagem resultante da adição
axs[2].imshow(added_img)
axs[2].set_title('Imagem Resultante')
axs[2].axis('off')
plt.show()
```

```python
print("Shape:", np.shape(img2), "Pixel Max:", img2.max())
```

```python
limiar = 0.9
# Criar uma máscara nas posições dos pixels que tem os pássaros
mask_bool = img2 < limiar
#print(mask_bool)

added_mask = img1
#print(added_mask[mask_bool])
#print(img2[mask_bool])

added_mask[mask_bool] = img2[mask_bool] * 0.8 # 0.8 corresponte a cor dos pássaros
# Multiplicar uma imagem por uma constante altera o 
#    contraste da imagem. Um valor maior aumenta o contraste, enquanto um valor menor o reduz.

plt.axis('off')
plt.imshow(added_mask)
```
