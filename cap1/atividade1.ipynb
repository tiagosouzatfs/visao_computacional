```python
import requests
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
link_image = "https://heltonmaia.com/computervision/_images/cover.jpeg"

image_bytes = requests.get(url=link_image)

type(image_bytes.content)
```

```python
# Transformar a imagem em bytes para Numpy array
img_array = np.frombuffer(image_bytes.content, dtype=np.uint8)
type(img_array)
#print(img_array)
```

```python
# Decodificar a imagem de bytes
# O cv2.IMREAD_COLOR  uma constante usada no OpenCV (biblioteca cv2) 
#    para especificar que a imagem deve ser lida em cores 
image_bgr = cv2.imdecode(img_array, cv2.IMREAD_COLOR)

plt.axis('off')
plt.imshow(image_bgr)
```

```python
# Verificando a cor em rgb
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

plt.axis('off')
plt.imshow(image_rgb)
```

```python
image_gray = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)

plt.axis('off')
plt.imshow(image_gray)
```

```python
image_gray.shape
```

```python
image_gray_text = cv2.putText(img=image_gray, text="OpenCV Challenge",
                              org=(400, 40), fontFace=cv2.FONT_HERSHEY_SIMPLEX,
                              fontScale=1.0, color=(255, 255, 255), thickness=2)

plt.axis('off')
plt.imshow(image_gray_text)
```

```python
# Salvando a imagem processada
cv2.imwrite('image_gray_text.png', image_gray_text)
```
