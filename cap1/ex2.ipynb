```python
import cv2
import matplotlib.pyplot as plt
```

```python
imagem = cv2.imread('image_rgb.png')

imagem_rgb = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagem)
plt.title('Imagem BGR')
plt.axis('off')
plt.show()
```

```python
plt.imshow(imagem_rgb)
plt.title('Imagem RGB')
plt.axis('off')
plt.show()
```
